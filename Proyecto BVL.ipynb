### Transformando los campos necesarios a tipo Numérico. Rellenando de 0 para los campos vacíos

```python
df['MontoNegociado']=pd.to_numeric(df.MontoNegociado, errors='coerce').fillna(0).astype(float)
df['CantidadNegociada']=pd.to_numeric(df.CantidadNegociada, errors='coerce').fillna(0).astype(float)
df['Apertura']=pd.to_numeric(df.Apertura, errors='coerce').fillna(0).astype(float)
df['Cierre']=pd.to_numeric(df.Cierre, errors='coerce').fillna(0).astype(float)
df['Maxima']=pd.to_numeric(df.Maxima, errors='coerce').fillna(0).astype(float)
df['Minima']=pd.to_numeric(df.Minima, errors='coerce').fillna(0).astype(float)
df['Promedio']=pd.to_numeric(df.Promedio, errors='coerce').fillna(0).astype(float)
df['CierreAnterior']=pd.to_numeric(df.CierreAnterior, errors='coerce').fillna(0).astype(float)
```

```python
df.dtypes
```

```python
## Este campo lo transformamos a Fecha
df['FechaAnterior']=pd.to_datetime(df.FechaAnterior, errors='coerce')
```

```python
## Revisamos nuevamente los primeros registros
df.head()
```

### Eliminamos las columnas que no vamos a utilizar

```python
del df['Id']
del df['FechaAnterior']
del df['CierreAnterior']
```
